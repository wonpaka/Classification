```python
# compared_models 함수를 통해 15개의 기본 모델을 학습하고 성능을 비교할 수 있습니다.
# F1 기준으로 성능이 가장 좋은 3개의 모델을 추려내어 저장해보겠습니다. 
# 본 대회 평가지표가 F1이기 때문에 F1 기준으로 모델을 선정합니다. # 따라서 F1을 기준으로 정렬함
best_3 = compare_models(sort = 'F1', n_select = 3)

# 상단 3가지 모델이 best_3에 저장됨
```

```python
# 보면 3가지 분류값이 들어있음
best_3
```

## 앙상블 모델 구축

```python
# 양상블 hard vs soft (다수결분류중 두가지로 나뉨)
# hard는 다수결투표 3가지 분류기법중 2:1로 결과가 나오면 2가나온것이 정답이됨
# soft는 분류의 결과가 2가지 일때 예측값의 평균을 내어 높은거로 최종예측함.

# 앙상블을 쓰는 이유 : 개별 모델의 성능이 안좋을 때 성능이 향상 / 과적합 방지
blended = blend_models(estimator_list = best_3, fold = 5, method = 'soft')
```

## 구축된 앙상블 모델로 예측

```python
# 이미 상단 setup에서 training set과 test set을 나눠두었다.
# 따라서 만들어진 모델로 test set을 돌려봄 (test 데이터와 다름)
# 주어진 train 데이터를 다시 한 번 train / validation으로 나눠서 실험을 한 것
pred_holdout = predict_model(blended)
```
